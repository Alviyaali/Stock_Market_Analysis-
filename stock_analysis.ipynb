```python
df_nifty['Date'] = pd.to_datetime(df_nifty['Date'], format='%d-%m-%Y')
df_hdfc['Date'] = pd.to_datetime(df_hdfc['Date'], format='%d-%m-%Y')
df_nifty = df_nifty.sort_values('Date')
df_hdfc = df_hdfc.sort_values('Date')
```

```python
#Nifty 50 stock showing closing price, open , high , low of last 30 days
plt.figure(figsize=(12,6))
plt.plot(df_nifty['Open'], label='Open', color='blue', linestyle='-', marker='o')
plt.plot(df_nifty['High'], label='High', color='green', linestyle='-', marker='^')
plt.plot(df_nifty['Low'], label='Low', color='red', linestyle='-', marker='v')
plt.plot(df_nifty['Price'], label='Closing Price', color='purple', linestyle='-', marker='x')
plt.title('Nifty 50: Open, High, Low, and Close Prices (Last 30 Days)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (INR)', fontsize=12)
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#HDFC stock showing closing price, open , high , low of last 30 days
plt.figure(figsize=(12,6))
plt.plot(df_hdfc['Open'], label='Open', color='blue', linestyle='-', marker='o')
plt.plot(df_hdfc['High'], label='High', color='green', linestyle='-', marker='^')
plt.plot(df_hdfc['Low'], label='Low', color='red', linestyle='-', marker='v')
plt.plot(df_hdfc['Price'], label='Closing Price', color='purple', linestyle='-', marker='x')
plt.title('HDFC: Open, High, Low, and Close Prices (Last 30 Days)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (INR)', fontsize=12)
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Closing price trend of Nifty 50 vs HDFC stock
plt.figure(figsize=(10,4))
plt.plot(df_nifty['Date'], df_nifty['Price'], label='Nifty 50 Closing Price', color='blue')
plt.plot(df_hdfc['Date'], df_hdfc['Price'], label='HDFC Closing Price', color='green')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing Price Trends: Nifty 50 vs HDFC')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#line chart showing historical volume of last 30 days of Nifty 50 vs HDFC
plt.figure(figsize=(20, 12))
plt.plot(df_nifty['Date'], df_nifty['Vol.'], label='Nifty 50 Volume', color='blue')
plt.plot(df_hdfc['Date'], df_hdfc['Vol.'], label='HDFC Volume', color='green')
plt.xlabel('Date')
plt.ylabel('Total Volume Traded')
plt.title('Historical Volume: Nifty 50 vs HDFC')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Histogram showing closing price distribution of Nifty 50 and HDFC
plt.figure(figsize=(14,6))

# Plot for Nifty 50
plt.subplot(1, 2, 1)
sns.histplot(df_nifty['Price'], bins=30, kde=True, color='blue')
plt.title('Nifty 50 Closing Price Distribution')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')

# Plot for HDFC
plt.subplot(1, 2, 2)
sns.histplot(df_hdfc['Price'], bins=30, kde=True, color='green')
plt.title('HDFC Closing Price Distribution')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

# Bar chart for Nifty 50 High prices
plt.subplot(2, 2, 1)
plt.bar(df_nifty['Date'], df_nifty['High'], color='skyblue')
plt.title('Nifty 50 High Prices')
plt.xlabel('Date')
plt.ylabel('High Price')
plt.xticks(rotation=45)

# Bar chart for Nifty 50 Low prices
plt.subplot(2, 2, 2)
plt.bar(df_nifty['Date'], df_nifty['Low'], color='lightcoral')
plt.title('Nifty 50 Low Prices')
plt.xlabel('Date')
plt.ylabel('Low Price')
plt.xticks(rotation=45)

# Bar chart for HDBK High prices
plt.subplot(2, 2, 3)
plt.bar(df_hdfc['Date'], df_hdfc['High'], color='lightgreen')
plt.title('HDFC High Prices')
plt.xlabel('Date')
plt.ylabel('High Price')
plt.xticks(rotation=45)

# Bar chart for HDFC Low prices
plt.subplot(2, 2, 4)
plt.bar(df_hdfc['Date'], df_hdfc['Low'], color='lightpink')
plt.title('HDFC Low Prices')
plt.xlabel('Date')
plt.ylabel('Low Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculated a 5-day moving average for a smoother trend line
df_nifty['Moving_Avg'] = df_nifty['Price'].rolling(window=5).mean()
df_hdfc['Moving_Avg'] = df_hdfc['Price'].rolling(window=5).mean()

fig, ax = plt.subplots(2, 1, figsize=(10, 6))

# Bar plot for overall performance change (first and last price comparison)
ax[0].bar(['Nifty 50', 'HDFC'], [df_nifty['Price'].iloc[-1] - df_nifty['Price'].iloc[0], 
                                 df_hdfc['Price'].iloc[-1] - df_hdfc['Price'].iloc[0]],
           color=['blue', 'green'])
ax[0].set_title('Overall Performance Change Over Time')
ax[0].set_ylabel('Change in Price')

# Line plot with moving averages for smoother trend visualization
ax[1].plot(df_nifty['Date'], df_nifty['Moving_Avg'], label='Nifty 50 (5-Day Avg)', color='blue', linestyle='--')
ax[1].plot(df_hdfc['Date'], df_hdfc['Moving_Avg'], label='HDFC (5-Day Avg)', color='green', linestyle='--')
ax[1].set_title('Nifty 50 vs HDFC - 5-Day Moving Average Trend')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Moving Average Price')
ax[1].legend()
ax[1].grid(visible=True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculated daily returns for both Nifty 50 and HDFC
df_nifty['Return'] = df_nifty['Price'].pct_change()
df_hdfc['Return'] = df_hdfc['Price'].pct_change()

# Calculated risk (standard deviation) and return (mean return)
nifty50_risk = df_nifty['Return'].std()
nifty50_return = df_nifty['Return'].mean()

hdfc_risk = df_hdfc['Return'].std()
hdfc_return = df_hdfc['Return'].mean()

# Created a DataFrame for plotting
risk_return_data = pd.DataFrame({
    'Asset': ['Nifty 50', 'HDFC'],
    'Risk': [nifty50_risk, hdfc_risk],
    'Return': [nifty50_return, hdfc_return]
})

# Plotted the Risk-Return Scatter Plot
plt.figure(figsize=(10, 4))
sns.scatterplot(data=risk_return_data, x='Risk', y='Return', hue='Asset', s=100)
plt.title('Risk-Return Scatter Plot')
plt.xlabel('Risk (Standard Deviation)')
plt.ylabel('Return (Average Return)')
plt.grid(True)
plt.show()
```

```python
# Visualization of the percent change for each day of Nifty 50 and HDFC stock
df_nifty['Daily Return'] = df_nifty['Price'].pct_change()
plt.grid(True)
df_nifty['Daily Return'].plot(figsize=(12,4),legend=True,linestyle='--',marker='o')
df_hdfc['Daily Return'] = df_hdfc['Price'].pct_change()
plt.grid(True)
df_hdfc['Daily Return'].plot(figsize=(12,4),legend=True,color="green", linestyle='--',marker='v')
```

```python
# Histogram to visualize the average daily return of Nifty 50 stock
sns.displot(df_nifty['Daily Return'].dropna(),bins=100,color='purple')
```

```python
# Histogram to visualize the average daily return of HDFC stock
sns.displot(df_hdfc['Daily Return'].dropna(),bins=100,color='green')
```
